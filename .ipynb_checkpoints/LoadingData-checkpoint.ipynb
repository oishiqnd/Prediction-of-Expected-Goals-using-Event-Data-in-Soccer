```python
from statsbombpy import sb
import pandas as pd
```

```python
sb.matches(competition_id = 16, season_id = 2)
```

```python
events = sb.events(match_id=18244)
```

```python
events.columns
```

```python
events = events[["id",'second','team', 'type', 'minute','pass_end_location', 'player', 'location', 'pass_outcome']]
events = events[events['team'] == 'Real Madrid'].reset_index()
```

```python
events.head(20)
```

```python
events['location'] = events['location'].astype(str).str.replace('[', '')
events['location'] = events['location'].astype(str).str.replace(']', '')

events['pass_end_location'] = events['pass_end_location'].astype(str).str.replace('[', '')
events['pass_end_location'] = events['pass_end_location'].astype(str).str.replace(']', '')

events[['endX', 'endY']] = events['pass_end_location'].str.split(',', 1, expand=True)
events[['X', 'Y']] = events['location'].str.split(',', 1, expand=True)

events = events.drop('pass_end_location', axis=1)
events = events.drop('location', axis=1)
```

```python
subs = events[events['type'] == 'Substitution']
subs= subs['minute']
firstSub = subs.min()
```

```python
events['passer'] = events['player']
events['recipient'] = events['player'].shift(-1)

passes = events[events['type'] == 'Pass']
passes.pass_outcome = passes.pass_outcome.fillna('Successful')
passes = passes[passes['pass_outcome'] == 'Successful']

passes = passes.drop('player', axis = 1)
passes = passes[passes['minute'] < firstSub]
```

```python
passes
```

```python
passes['X'] = pd.to_numeric(passes['X'], downcast = 'float')
passes['Y'] = pd.to_numeric(passes['Y'], downcast = 'float')
passes['endX'] = pd.to_numeric(passes['endX'], downcast = 'float')
passes['endY'] = pd.to_numeric(passes['endY'], downcast = 'float')
```

```python
average_locations =  passes.groupby('passer').agg({'X':['mean'], 'Y':['mean', 'count']})
average_locations.columns = ['x','y','count']
```

```python
average_locations
```

```python
pass_between = passes.groupby(['passer', 'recipient']).id.count().reset_index()
```

```python
pass_between.rename({'id':'pass_count'}, axis = 'columns', inplace = True)
```

```python
pass_between = pass_between.merge(average_locations, left_on= 'passer', right_index=True)
pass_between = pass_between.merge(average_locations, left_on= 'recipient', right_index=True, suffixes=['','_end'])
```

```python
pass_between
```

```python
pass_between = pass_between[pass_between['pass_count'] >3]
```

```python
pass_between
```

```python
from mplsoccer import Pitch
import matplotlib.pyplot as plt
pitch = Pitch()
pitch = Pitch(pitch_color='#212326', line_color='white')
# specifying figure size (width, height)

fig, ax = pitch.draw(figsize=(10, 5))
arrows = pitch.arrows(1.2*pass_between.x,.8*pass_between.y,1.2*pass_between.x_end,.8*pass_between.y_end, ax=ax,
                     width = 3, headwidth = 3, color='white', zorder=1, alpha = .5)

nodes = pitch.scatter(1.2*average_locations.x,.8*average_locations.y,
                     s=300, color= '#d3d3d3', edgecolor = 'black', linewidth = 2.5, alpha =1, zorder=1, ax=ax)
```
